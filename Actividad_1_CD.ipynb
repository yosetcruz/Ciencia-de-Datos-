```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_Agua_Santa = pd.read_csv(
    "Datos SINAICA - Agua Santa - PM10 - PM24H - 2025-01-01 - 2026-01-01.csv", encoding='latin-1')
```

```python
df_Agua_Santa
```

```python
df_Agua_Santa.dropna(inplace=True)
df_Agua_Santa
```

```python
df_Agua_Santa["Promedio mÃ³vil 24 horas"]
```

```python
pm10 = pd.to_numeric(df_Agua_Santa["Promedio mÃ³vil 24 horas"], errors='coerce').to_numpy()

limite_oms = 45
excede = pm10 > limite_oms

fechas = df_Agua_Santa["Fecha"].to_numpy()
dias_excedidos = fechas[excede]

print(dias_excedidos)
print(len(dias_excedidos))
```

```python
pm10 = df_Agua_Santa["Promedio mÃ³vil 24 horas"].to_numpy()
fechas = df_Agua_Santa["Fecha"].to_numpy()

limite_oms = 45
exceso = pm10 > limite_oms

plt.figure(figsize=(30,10))

plt.plot(fechas, pm10)
plt.axhline(limite_oms, color='red', linestyle='--')

plt.scatter(fechas[exceso], pm10[exceso], color='red')

step = 50  # â muestra una fecha cada 50 puntos
plt.xticks(fechas[::step], rotation=45)

plt.tight_layout()
plt.show()
```
